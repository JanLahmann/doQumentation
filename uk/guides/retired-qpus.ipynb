## Виведені з експлуатації сімейства процесорів
- **Canary**: Сімейство Canary складалося з невеликих конструкцій, що містили від 5 до 16 кубітів.
- **Falcon**: Сімейство пристроїв Falcon забезпечувало цінну платформу для схем середнього масштабу, а також слугувало платформою для демонстрації вдосконалень продуктивності та масштабованості перед їх переносом на більші пристрої.
- **Egret**: Egret впровадив інновації у вигляді перестроюваних зв'язувачів на 33-кубітній платформі, що забезпечило більш швидкі та точні двокубітні вентилі.
- **Hummingbird**: Використовуючи розташування кубітів у формі важкого шестикутника, сімейство Hummingbird підтримувало до 65 кубітів.
- **Osprey**: Osprey був майже вчетверо більшим за Eagle — 433 кубіти. Більший розмір чіпа вимагав додаткових удосконалень у корпусуванні пристрою, а також спеціальних гнучких кабелів у кріостаті для задоволення більших вимог введення/виведення в межах тієї самої трасування проводки.
- **Eagle**: Сімейство процесорів Eagle використовувало більш масштабовані технології пакування, ніж попередні покоління. Зокрема, сигнали проходили через кілька шарів чіпа, що забезпечувало щільне введення/виведення без втрати продуктивності.
