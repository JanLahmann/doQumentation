## Définir les options Executor
Si une option est spécifiée à la fois pendant et après l'initialisation de la primitive, la valeur définie après l'initialisation de la primitive est utilisée.

### Initialisation de la primitive
Tu peux passer une instance de la classe d'options ou un dictionnaire lors de l'initialisation d'Executor, qui fait ensuite une copie de ces options. Ainsi, la modification du dictionnaire original ou de l'instance d'options n'affecte pas les options appartenant à la primitive.

#### Classe d'options
Lors de la création d'une instance de la classe `Executor`, tu peux passer une instance de la classe d'options. Ces options sont ensuite appliquées lorsque tu utilises `run()` pour effectuer le calcul. Spécifie les options dans ce format : `options.option.sous-option.sous-sous-option = choix`. Par exemple : `options.environment.log_level = INFO`.

Exemple :

````python
from qiskit_ibm_runtime import QiskitRuntimeService, Executor
from qiskit_ibm_runtime.options import ExecutorOptions

service = QiskitRuntimeService()
backend = service.least_busy(operational=True, simulator=False)

options = ExecutorOptions(
    environment={"log_level": "INFO"},
    execution={"init_qubits": True},
)

# or use the following instead:

options = ExecutorOptions()
options.environment.log_level = "INFO"
options.execution.init_qubits = True

executor = Executor(mode=backend, options=options)
````

#### Dictionnaire
Tu peux spécifier les options sous forme de dictionnaire lors de l'initialisation d'Executor.

````python
from qiskit_ibm_runtime import QiskitRuntimeService, Executor

service = QiskitRuntimeService()
backend = service.least_busy(operational=True, simulator=False)

# Setting options during primitive initialization
executor = Executor(
    backend,
    options={
        "environment": {"log_level": "INFO"},
        "execution": {"init_qubits": True},
    },
)
````

## Compatibilité des fonctionnalités
Certaines fonctionnalités d'exécution ne peuvent pas être utilisées ensemble dans un seul job. Clique sur l'onglet approprié pour obtenir une liste des fonctionnalités incompatibles avec la fonctionnalité sélectionnée :

<Tabs>
  <TabItem value="Dynamic" label="Circuits dynamiques">
  Incompatible avec :
  - ZNE par repliement de gate
  - PEA
  - PEC
  - Découplage dynamique

  Autres notes :
  - La torsion de gate peut être appliquée aux circuits dynamiques, mais uniquement aux gates qui ne se trouvent pas dans des blocs conditionnels. La torsion de mesure ne peut être appliquée qu'aux mesures terminales.
  - Compatible avec les gates fractionnaires lors de l'utilisation de `qiskit-ibm-runtime` v0.42.0 ou ultérieur.

  </TabItem>
  <TabItem value="DD" label="Découplage dynamique">
  Incompatible avec les circuits dynamiques.

  </TabItem>

  <TabItem value="Fractional" label="Gates fractionnaires">
  Incompatible avec :
  - Torsion de gate
  - PEA
  - PEC

  Compatible avec les circuits dynamiques lors de l'utilisation de `qiskit-ibm-runtime` v0.42.0 ou ultérieur.

  </TabItem>

  <TabItem value="ZNE" label="ZNE par repliement de gate">
    Incompatible avec :
  - Circuits dynamiques
  - PEA
  - PEC

  Pourrait ne pas fonctionner avec des gates personnalisées.
  </TabItem>
  <TabItem value="Twirling" label="Torsion de gate">
  Incompatible avec les gates fractionnaires ou les étirements.

  Autres notes :
  - La torsion de gate peut être appliquée aux circuits dynamiques, mais uniquement aux gates qui ne se trouvent pas dans des blocs conditionnels. La torsion de mesure ne peut être appliquée qu'aux mesures terminales.
  - Ne fonctionne pas avec des enchevêtreurs non Clifford.

  </TabItem>

  <TabItem value="PEA" label="PEA">
    Incompatible avec :
  - Circuits dynamiques
  - Gates fractionnaires
  - ZNE par repliement de gate
  - PEC
  </TabItem>

  <TabItem value="PEC" label="PEC">

  - Circuits dynamiques
