Penekanan error mengacu pada teknik-teknik di mana kamu menggunakan pengetahuan tentang efek yang tidak diinginkan untuk memperkenalkan kustomisasi yang bisa mengantisipasi dan menghindari dampak potensial dari efek tersebut. Teknik-teknik ini seringkali terdiri dari mengubah atau menambahkan sinyal kontrol untuk memastikan prosesor kuantum mengembalikan hasil yang diinginkan. Hal ini biasanya menghasilkan overhead pra-pemrosesan kuantum; oleh karena itu, penting untuk mencapai keseimbangan antara menyempurnakan hasil kamu dan memastikan pekerjaan kamu selesai dalam waktu yang wajar.

Primitif mendukung sejumlah teknik penekanan error, termasuk [dynamical decoupling](https://docs.quantum.ibm.com/api/qiskit-ibm-runtime/options-dynamical-decoupling-options#dynamicaldecouplingoptions) dan [Pauli twirling](https://docs.quantum.ibm.com/api/qiskit-ibm-runtime/options-twirling-options). Lihat [Teknik mitigasi dan penekanan error](error-mitigation-and-suppression-techniques) untuk penjelasan masing-masing. Saat menggunakan primitif, kamu bisa mengaktifkan atau menonaktifkan metode individual. Lihat bagian [Opsi penekanan error lanjutan](#transpilation-table) untuk detailnya.

Estimator menggunakan penekanan dan mitigasi error secara default. Jika kamu tidak ingin ada pemrosesan yang dilakukan pada Circuit input kamu, ikuti instruksi di bagian [Matikan semua mitigasi error dan penekanan error](specify-runtime-options#no-error-mitigation).

## Opsi penekanan error lanjutan
Dalam primitif, kamu bisa secara eksplisit mengaktifkan dan menonaktifkan metode mitigasi dan penekanan error individual, seperti dynamical decoupling.

> **Note:** - Tidak semua opsi tersedia untuk kedua primitif. Lihat tabel [opsi yang tersedia](runtime-options-overview#options-table) untuk daftar opsi yang tersedia.
> - Tidak semua metode bekerja bersama pada semua jenis Circuit. Lihat tabel [kompatibilitas fitur](runtime-options-overview#options-compatibility-table) untuk detailnya.

````python
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit_ibm_runtime import SamplerV2 as Sampler

service = QiskitRuntimeService()
backend = service.least_busy(operational=True, simulator=False)

sampler = Sampler(backend)

# Turn on dynamical decoupling with sequence XpXm.
sampler.options.dynamical_decoupling.enable = True
sampler.options.dynamical_decoupling.sequence_type = "XpXm"

print(
    f">>> dynamical decoupling sequence to use: {sampler.options.dynamical_decoupling.sequence_type}"
)
````
