Selain [memvisualisasikan instruksi pada Circuit](/guides/visualize-circuits), kamu mungkin juga ingin memvisualisasikan penjadwalan Circuit menggunakan metode [`timeline_drawer`](https://docs.quantum.ibm.com/api/qiskit/qiskit.visualization.timeline_drawer) dari Qiskit. Visualisasi ini bisa membantu kamu dengan cepat melihat waktu idle pada Qubit, misalnya. Namun, metode ini tidak memberikan hasil yang akurat untuk dynamic circuits. Untuk memvisualisasikan penjadwalan dynamic circuit, gunakan metode `draw_circuit_schedule_timing`, seperti yang dijelaskan di bagian [Dukungan Qiskit Runtime](#qr-support).

## Contoh

Untuk memvisualisasikan program Circuit terjadwal, kamu bisa memanggil fungsi ini dengan sekumpulan argumen kontrol. Sebagian besar tampilan gambar output bisa dimodifikasi melalui stylesheet, tapi ini tidak wajib.

### Gambar dengan stylesheet default

````python
from qiskit import QuantumCircuit
from qiskit.visualization.timeline import draw
from qiskit.providers.fake_provider import GenericBackendV2
from qiskit.transpiler import generate_preset_pass_manager

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)

backend = GenericBackendV2(5)

pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_circuit = pm.run(qc)

draw(isa_circuit, target=backend.target)
````

![Output dari sel kode sebelumnya](../docs/images/guides/visualize-circuit-timing/extracted-outputs/5b21b8c1-0.svg)

### Gambar dengan stylesheet untuk debugging program

````python
from qiskit import QuantumCircuit
from qiskit.visualization.timeline import draw, IQXDebugging
from qiskit.providers.fake_provider import GenericBackendV2
from qiskit.transpiler import generate_preset_pass_manager

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

backend = GenericBackendV2(5)
pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_circuit = pm.run(qc)
draw(isa_circuit, style=IQXDebugging(), target=backend.target)
````

![Output dari sel kode sebelumnya](../docs/images/guides/visualize-circuit-timing/extracted-outputs/907dc46c-0.svg)

Kamu bisa membuat fungsi generator atau layout kustom dan memperbarui stylesheet yang sudah ada dengan fungsi kustom tersebut. Dengan cara ini, kamu bisa mengontrol sebagian besar tampilan gambar output tanpa memodifikasi codebase dari drawer Circuit terjadwal. Lihat referensi API [`timeline_drawer`](https://docs.quantum.ibm.com/api/qiskit/qiskit.visualization.timeline_drawer) untuk contoh lebih lanjut.

## Dukungan Qiskit Runtime
Meskipun timeline drawer yang ada di Qiskit berguna untuk static circuit, ia mungkin tidak secara akurat mencerminkan timing [dynamic circuits](/guides/classical-feedforward-and-control-flow) karena adanya operasi implisit seperti broadcasting dan penentuan cabang. Sebagai bagian dari dukungan dynamic circuit, Qiskit Runtime mengembalikan informasi timing Circuit yang akurat di dalam hasil job bila diminta.

> **Note:** - Ini adalah fungsi eksperimental. Fungsi ini masih dalam status preview release dan oleh karena itu dapat berubah sewaktu-waktu.
> - Fungsi ini hanya berlaku untuk job Qiskit Runtime Sampler.
> - Meskipun total waktu Circuit dikembalikan dalam metadata "compilation", ini BUKAN waktu yang digunakan untuk penagihan (quantum time).
### Aktifkan pengambilan data timing
Untuk mengaktifkan pengambilan data timing, atur flag eksperimental `scheduler_timing` ke `True` saat menjalankan job primitive.
