## Pangkalahatang-ideya
Ginagawa ng Fire Opal Performance Management na simple para sa sinuman na makamit ang makabuluhang resulta mula sa mga quantum computer sa malaking sukat nang hindi kailangang maging eksperto sa quantum hardware. Kapag nagpapatakbo ng mga circuit gamit ang Fire Opal Performance Management, awtomatikong inilalapat ang mga AI-driven na teknik ng error suppression, na nagbibigay-daan sa pag-scale ng mas malalaking problema na may mas maraming gate at qubit. Binabawasan ng approach na ito ang bilang ng mga shot na kailangan para makamit ang tamang sagot, nang walang dagdag na overhead — na nagresulta sa malaking tipid sa oras ng computation at gastos.

Sinusupress ng Performance Management ang mga error at pinapataas ang posibilidad na makuha ang tamang sagot sa maingay na hardware. Sa madaling salita, pinapataas nito ang signal-to-noise ratio. Ipinapakita ng sumusunod na larawan kung paano maaaring bawasan ng mas mataas na katumpakan na pinapagana ng Performance Management ang pangangailangan para sa karagdagang mga shot sa kaso ng isang 10-qubit Quantum Fourier Transform algorithm. Sa 30 shot lamang, naabot ng Q-CTRL ang 99% confidence threshold, samantalang ang default (`QiskitRuntime` Sampler, `optimization_level`=3 at `resilience_level`=1, `ibm_sherbrooke`) ay nangangailangan ng 170,000 shot. Sa pamamagitan ng pagkuha ng tamang sagot nang mas mabilis, makakatipid ka ng malaking oras ng computation.

![Visualization ng pinahusay na runtime](../docs/images/guides/qctrl-performance-management/achieve_more.svg)

Maaaring gamitin ang Performance Management function sa anumang algorithm, at madali mo itong magagamit bilang kapalit ng mga standard na [Qiskit Runtime primitives](/guides/primitives). Sa likod ng eksena, maraming teknik ng error suppression ang nagtatrabaho nang sabay-sabay upang maiwasan ang mga error sa runtime. Lahat ng Fire Opal pipeline method ay pre-configured at algorithm-agnostic, ibig sabihin palagi kang makakakuha ng pinakamahusay na performance nang walang karagdagang configuration.

Para makakuha ng access sa Performance Management, [makipag-ugnayan sa Q-CTRL](https://form.typeform.com/to/uOAVDnGg?typeform-source=q-ctrl.com).

## Paglalarawan
Ang Fire Opal Performance Management ay may dalawang opsyon para sa pagpapatakbo na katulad ng mga Qiskit Runtime primitives, kaya madali mong mapapalitan ng Q-CTRL Sampler at Estimator. Ang pangkalahatang workflow para sa paggamit ng Performance Management function ay:
1. Tukuyin ang iyong circuit (at mga operator sa kaso ng Estimator).
2. Patakbuhin ang circuit.
3. Kunin ang mga resulta.

Para bawasan ang ingay ng hardware, gumagamit ang Fire Opal ng iba't ibang AI-driven na teknik ng error suppression na ipinapakita sa sumusunod na larawan. Sa Fire Opal, ang buong pipeline ay ganap na automated nang walang pangangailangan para sa configuration.

Inaalis ng pipeline ng Fire Opal ang pangangailangan para sa karagdagang overhead, tulad ng pinahusay na quantum runtime o dagdag na pisikal na qubit. Tandaan na ang classical processing time ay nananatiling isang salik (sumangguni sa seksyon ng [Mga Benchmark](#benchmarks) para sa mga pagtatantya, kung saan ang "Total time" ay sumasalamin sa parehong classical at quantum processing). Kumpara sa error mitigation, na nangangailangan ng overhead sa anyo ng sampling, ang error suppression ng Fire Opal ay gumagawa sa parehong gate at pulse na antas upang tugunan ang iba't ibang pinagkukunan ng ingay at upang maiwasan ang posibilidad ng paglitaw ng error. Sa pamamagitan ng pag-iwas sa mga error, naaalis ang pangangailangan para sa mahal na post-processing.

Ipinapakita ng sumusunod na larawan ang mga paraan ng error suppression na na-automate ng Fire Opal Performance Management.

![Visualization ng error suppression pipeline](../docs/images/guides/qctrl-performance-management/error_suppression.svg)

Nag-aalok ang function ng dalawang primitive, Sampler at Estimator, at ang mga input at output ng pareho ay nagpapalawak ng implemented spec para sa [Qiskit Runtime V2 primitives](/guides/primitive-input-output#pubs).

## Mga Benchmark
Ang mga resulta ng [nai-publish na algorithmic benchmarking](https://journals.aps.org/prapplied/abstract/10.1103/PhysRevApplied.20.024034) ay nagpapakita ng makabuluhang pagpapabuti ng performance sa iba't ibang algorithm, kabilang ang Bernstein-Vazirani, quantum Fourier transform, Grover's search, quantum approximate optimization algorithm, at variational quantum eigensolver. Ang natitirang bahagi ng seksyong ito ay nagbibigay ng karagdagang detalye tungkol sa mga uri ng algorithm na maaari mong patakbuhin, pati na rin ang inaasahang performance at mga runtime.

Ang mga sumusunod na independyenteng pag-aaral ay nagpapakita kung paano pinapagana ng Performance Management ng Q-CTRL ang algorithmic na pananaliksik sa record-breaking na sukat:
- [Parametrized Energy-Efficient Quantum Kernels for Network Service Fault Diagnosis](https://arxiv.org/abs/2405.09724v1) - hanggang 50-qubit quantum kernel learning
- [Tensor-based quantum phase difference estimation for large-scale demonstration](https://arxiv.org/abs/2408.04946) - hanggang 33-qubit quantum phase estimation
- [Hierarchical Learning for Quantum ML: Novel Training Technique for Large-Scale Variational Quantum Circuits](https://arxiv.org/abs/2311.12929) - hanggang 21-qubit quantum data loading

Ang sumusunod na talahanayan ay nagbibigay ng magaspang na gabay sa katumpakan at mga runtime mula sa mga nakaraang benchmarking run sa `ibm_fez`. Ang performance sa ibang device ay maaaring magkaiba. Ang oras ng paggamit ay batay sa assumption na 10,000 shot bawat circuit. Ang "Bilang ng mga qubit" na ipinahiwatig ay hindi isang mahirap na limitasyon kundi kumakatawan sa magaspang na mga threshold kung saan maaari kang umasa ng napaka-consistent na katumpakan ng solusyon. Matagumpay na nalutas ang mas malalaking sukat ng problema, at hinihikayat ang pagsubok na higit pa sa mga limitasyong ito.

| Halimbawa    | Bilang ng qubit | Katumpakan | Sukatan ng katumpakan | Kabuuang oras (s) | Paggamit ng runtime (s) | Primitive (Mode) |
| ---------  | ---------------- | -------------------------- | -------- | ---------- | ------------- |------------- |
| Bernstein–Vazirani  |  50Q    | 100%  | Success Rate (Porsyento ng mga run kung saan ang tamang sagot ay ang pinakamataas na bilang ng bitstring)     | 10    | 8         | Sampler |
| Quantum Fourier Transform | 30Q              | 100% | Success Rate (Porsyento ng mga run kung saan ang tamang sagot ay ang pinakamataas na bilang ng bitstring)      | 10    | 8        | Sampler |
| Quantum Phase Estimation  | 30Q   | 99.9998%  | Katumpakan ng nahanap na anggulo: `1- abs(real_angle - angle_found)/pi`  | 10  | 8  | Sampler |
| Quantum simulation: Ising model (15 hakbang) | 20Q  | 99.775%   |  $A$ (tinukoy sa ibaba)  |  60 (bawat hakbang)  | 15 (bawat hakbang)   | Estimator |
| Quantum simulation 2: molecular dynamics (20 time point) | 34Q  |  96.78%  |  $A_{mean}$ (tinukoy sa ibaba)   | 10 (bawat time point)    | 6 (bawat time point)  | Estimator |

Pagtukoy ng katumpakan ng pagsukat ng expectation value — ang metric na $A$ ay tinukoy tulad ng sumusunod:
$$
A = 1 - \frac{|\epsilon^{ideal} - \epsilon^{meas}|}{\epsilon^{ideal}_{max} - \epsilon^{ideal}_{min}},
$$
kung saan ang $ \epsilon^{ideal} $ = ideal na expectation value, $ \epsilon^{meas} $ = nasukat na expectation value, $\epsilon^{ideal}_{max} $ = ideal na maximum value, at $\epsilon^{ideal}_{min}$ = ideal na minimum value. Ang $A_{mean}$ ay simpleng average ng halaga ng $A$ sa maraming pagsukat.

Ginagamit ang metric na ito dahil ito ay invariant sa mga global shift at scaling sa hanay ng mga maaabot na halaga. Sa madaling salita, anuman ang itaas o ibaba mo ang hanay ng mga posibleng expectation value o palakihin ang kalat, dapat manatiling consistent ang halaga ng $A$.

## Estimator primitive

### Estimator example

Use Fire Opal Performance Management's Estimator primitive to determine the expectation value of a single circuit-observable pair.

In addition to the `qiskit-ibm-catalog` and `qiskit` packages, you will also use the `numpy` package to run this example. You can install this package by uncommenting the following cell if you are running this example in a notebook using the IPython kernel.

````python
# %pip install numpy
````

**2. Patakbuhin ang circuit**

Patakbuhin ang circuit at opsyonal na tukuyin ang backend at bilang ng mga shot.

````python
import numpy as np
from qiskit.circuit.library import iqp
from qiskit.quantum_info import random_hermitian, SparsePauliOp

n_qubits = 50

# Generate a random circuit
mat = np.real(random_hermitian(n_qubits, seed=1234))
circuit = iqp(mat)
circuit.measure_all()

# Define observables as a string
observable = SparsePauliOp("Z" * n_qubits)
````

````python
# Create PUB tuple
estimator_pubs = [(circuit, observable)]
````

Maaari mong gamitin ang mga pamilyar na [Qiskit Serverless APIs](/guides/serverless) para suriin ang status ng iyong Qiskit Function workload:

````python
# This cell is hidden from users
from qiskit_ibm_runtime import QiskitRuntimeService

service = QiskitRuntimeService()
backend_name = service.least_busy().name
````

````python
# Run the circuit using Estimator
qctrl_estimator_job = perf_mgmt.run(
    primitive="estimator",
    pubs=estimator_pubs,
    backend_name=backend_name,
)
````

**3. Kunin ang resulta**

````python
qctrl_estimator_job.status()
````

Ang mga resulta ay may parehong format tulad ng isang [Estimator result](/guides/estimator-input-output):

````python
# Retrieve the counts from the result list
result = qctrl_estimator_job.result()
````

The results have the same format as an [Estimator result](/docs/guides/estimator-input-output):

````python
import numpy

result_str = str(result)

with numpy.printoptions(threshold=200):
    print(
        f"The result of the submitted job had {len(result)} PUB "
        f"and has a value:\n {result[0]}\n"
    )

print("The associated PubResult of this job has the following DataBins:")
print(f"{result[0].data}\n")

print(f"And this DataBin has attributes: {result[0].data.keys()}")

print("The expectation values measured from this PUB are:")
print(f"{result[0].data.evs}")
````

## Sampler primitive
### Halimbawa ng Sampler
Gamitin ang Sampler primitive ng Fire Opal Performance Management para patakbuhin ang isang Bernstein–Vazirani circuit. Ang algorithm na ito, na ginagamit para hanapin ang isang nakatagong string mula sa mga output ng isang black box function, ay isang karaniwang benchmarking algorithm dahil mayroon itong isang tamang sagot.

**1. Gumawa ng circuit**

Tukuyin ang tamang sagot sa algorithm, ang nakatagong bitstring, at ang Bernstein–Vazirani circuit. Maaari mong ayusin ang lapad ng circuit sa pamamagitan lamang ng pagbabago ng `circuit_width`.

````python
import qiskit

circuit_width = 35
hidden_bitstring = "1" * circuit_width

# Create circuit, reserving one qubit for BV oracle
bv_circuit = qiskit.QuantumCircuit(circuit_width + 1, circuit_width)
bv_circuit.x(circuit_width)
bv_circuit.h(range(circuit_width + 1))
for input_qubit, bit in enumerate(reversed(hidden_bitstring)):
    if bit == "1":
        bv_circuit.cx(input_qubit, circuit_width)
bv_circuit.barrier()
bv_circuit.h(range(circuit_width + 1))
bv_circuit.barrier()
for input_qubit in range(circuit_width):
    bv_circuit.measure(input_qubit, input_qubit)

# Create PUB tuple
sampler_pubs = [(bv_circuit,)]
````

**2. Patakbuhin ang circuit**

Patakbuhin ang circuit at opsyonal na tukuyin ang backend at bilang ng mga shot.

````python
# Run the circuit using Sampler
qctrl_sampler_job = perf_mgmt.run(
    primitive="sampler",
    pubs=sampler_pubs,
    backend_name=backend_name,
)
````

Suriin ang [status](/guides/functions#check-job-status) ng iyong Qiskit Function workload o kunin ang [mga resulta](/guides/functions#retrieve-results) tulad ng sumusunod:

````python
# Print the ID so you can use it later, if necessary
print(qctrl_sampler_job.job_id)

qctrl_sampler_job.status()
````

**3. Retrieve the result**

````python
# Retrieve the job results
sampler_result = qctrl_sampler_job.result()
````

````python
# Get results for the first (and only) PUB
pub_result = sampler_result[0]
counts = pub_result.data.c.get_counts()

print("Counts for the meas output register (limited to 30 results):")
for i, (bitstring, count) in enumerate(counts.items()):
    if i >= 50:
        print(f"  ... ({len(counts) - 30} more items)")
        break
    print(f"  {bitstring}: {count}")
````

**3. Plot the top bitstrings**

Plot the bitstring with the highest counts to see if the hidden bitstring was the mode.

````python
import matplotlib.pyplot as plt


def plot_top_bitstrings(counts_dict, hidden_bitstring=None):
    # Sort and take the top 100 bitstrings
    top_100 = sorted(counts_dict.items(), key=lambda x: x[1], reverse=True)[
        :100
    ]
    if not top_100:
        print("No bitstrings found in the input dictionary.")
        return

    # Unzip the bitstrings and their counts
    bitstrings, counts = zip(*top_100)

    # Assign colors: purple if the bitstring matches hidden_bitstring,
    # otherwise gray
    colors = [
        "#680CE9" if bit == hidden_bitstring else "gray" for bit in bitstrings
    ]

    # Create the bar plot
    plt.figure(figsize=(15, 8))
    plt.bar(
        range(len(bitstrings)), counts, tick_label=bitstrings, color=colors
    )

    # Rotate the bitstrings for better readability
    plt.xticks(rotation=90, fontsize=8)
    plt.xlabel("Bitstrings")
    plt.ylabel("Counts")
    plt.title("Top 100 Bitstrings by Counts")

    # Show the plot
    plt.tight_layout()
    plt.show()
````

The hidden bitstring is highlighted in purple, and it should be the bitstring with the highest number of counts.

````python
plot_top_bitstrings(counts, hidden_bitstring)
````
