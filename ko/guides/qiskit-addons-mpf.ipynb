다중 곱 공식(MPF)은 Circuit 실행 횟수가 증가하는 대신 양자 시스템의 동역학을 더 정확하게 시뮬레이션하는 데 사용할 수 있습니다. 이것은 시간 진화된 상태에 대한 기댓값의 오류를 완화하는 후처리 기법입니다.

고전 컴퓨팅은 여러 Circuit 실행의 가중 조합에 계수를 제공하는 선형 방정식 시스템을 푸는 데 사용됩니다. 이 가중 조합을 사용하면 좋은 Trotter 단계 선택을 통해 시간 진화 시뮬레이션과 관련된 오류를 줄일 수 있습니다. MPF 도구는 Trotter 단계의 수와 Trotter 근사 차수를 포함한 선택된 데이터를 수집하여 관련 선형 방정식 시스템을 준비하고 풀거나 근사 해를 구합니다. 이를 통해 시간 진화된 상태의 기댓값 측정을 후처리하는 데 활용할 수 있습니다.
## MPF 패키지 설치
MPF 패키지를 설치하는 방법은 두 가지가 있습니다. PyPI를 통한 설치와 소스에서 빌드하는 방법입니다. 패키지 의존성 간의 분리를 보장하기 위해 [가상 환경](https://docs.python.org/3.10/tutorial/venv.html)에 설치하는 것을 권장합니다.
