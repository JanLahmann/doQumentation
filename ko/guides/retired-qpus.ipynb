## 은퇴한 프로세서 패밀리
- **Canary**: Canary 패밀리는 5개에서 16개의 Qubit을 포함하는 소형 설계로 구성되어 있습니다.
- **Falcon**: Falcon 패밀리 장치는 중간 규모 Circuit을 위한 유용한 플랫폼을 제공하였으며, 더 큰 장치에 적용하기 전에 성능 및 확장성 개선을 시연하는 플랫폼으로도 활용되었습니다.
- **Egret**: Egret은 33-Qubit 플랫폼에 조정 가능한 커플러(tunable coupler)의 혁신을 도입하여, 더 빠르고 높은 충실도의 2-Qubit Gate를 구현하였습니다.
- **Hummingbird**: 헤비-헥사고날(heavy-hexagonal) Qubit 레이아웃을 사용하여, Hummingbird 패밀리는 최대 65개의 Qubit을 지원하였습니다.
- **Osprey**: Osprey는 433 Qubit으로 Eagle의 거의 4배 크기였습니다. 더 큰 칩 크기에는 장치 패키징의 추가적인 개선이 필요하였으며, 동일한 배선 공간 내에서 더 많은 I/O 요구 사항을 수용하기 위해 극저온 장치(cryostat)에 맞춤형 플렉스 케이블이 사용되었습니다.
- **Eagle**: Eagle 프로세서 패밀리는 이전 세대보다 더 확장 가능한 패키징 기술을 도입하였습니다. 특히 신호가 여러 칩 레이어를 통과함으로써, 성능을 저하시키지 않고 고밀도 I/O가 가능해졌습니다.
