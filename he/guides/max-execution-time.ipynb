## הגדרת זמן ביצוע מקסימלי
זמן הביצוע המקסימלי למשימה הוא הקטן מבין הערכים הבאים:

- הערך שהוגדר עבור `max_execution_time`
- ערך פסק הזמן של המשימה הנקבע על ידי השירות

ערך `max_execution_time` מבוסס על *שימוש ב-QPU*, לא על זמן שעון קיר.

הגדר את זמן הביצוע המקסימלי (בשניות) באפשרויות המשימה, כפי שמוצג בדוגמה הבאה. ראה את המדריכים [אפשרויות Estimator](/guides/estimator-options), [אפשרויות Sampler](/guides/sampler-options), או [אפשרויות Executor](/guides/executor-options) לקבלת מידע על הגדרת אפשרויות.
