# זמן ביצוע מקסימלי לעומסי עבודה של Qiskit Runtime

כדי להבטיח הוגנות, קיים [זמן מקסימלי המחושב על ידי השירות](#max-QPU) שכל משימת Qiskit Runtime יכולה לרוץ. אם משימה חורגת ממגבלת זמן זו, היא מבוטלת בכפייה ומוגרעת חריגה מסוג `RuntimeJobMaxTimeoutError`. באפשרותך גם [להגדיר זמן ביצוע מקסימלי](#max-job) על המשימות שלך כדי לסייע בשליטה על עלויות.

 > **Note:** הסשן או ה-batch מסתיימים מיידית כאשר מגבלת העלות מגיעה לסיומה.

## הגדרת זמן ביצוע מקסימלי
זמן הביצוע המקסימלי למשימה הוא הקטן מבין הערכים הבאים:

- הערך שהוגדר עבור `max_execution_time`
- ערך פסק הזמן של המשימה הנקבע על ידי השירות

ערך `max_execution_time` מבוסס על _זמן קוונטי_, לא על זמן שעון קיר. זמן קוונטי הוא כמות הזמן שה-QPU מוקדש לעיבוד המשימה שלך.

הגדר את זמן הביצוע המקסימלי (בשניות) באפשרויות המשימה, כפי שמוצג בדוגמה הבאה. ראה [ציון אפשרויות](/guides/specify-runtime-options) לקבלת מידע על הגדרת אפשרויות.
