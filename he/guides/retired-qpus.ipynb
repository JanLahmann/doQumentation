## משפחות מעבדים שפרשו
- **Canary**: משפחת Canary כללה עיצובים קטנים המכילים בין 5 ל-16 qubits.
- **Falcon**: משפחת מכשירי Falcon הציעה פלטפורמה ערכית עבור Circuit-ים בקנה מידה בינוני, ושימשה גם כפלטפורמה יקרת ערך להדגמת שיפורי ביצועים ומדרגיות לפני שהועברו למכשירים הגדולים יותר.
- **Egret**: Egret הביא את החידושים של couplers כוונניים לפלטפורמת 33 qubits, מה שהוביל ל-Gate-ים בין שני qubits מהירים יותר ובעלי נאמנות גבוהה יותר.
- **Hummingbird**: באמצעות פריסת qubit משושה-כבד, משפחת Hummingbird אפשרה עד 65 qubits.
- **Osprey**: Osprey היה כמעט פי ארבעה מגודל Eagle עם 433 qubits. גדלי השבב הגדולים יותר דרשו שיפורים נוספים באריזת המכשיר, כמו גם כבלים גמישים מותאמים בקריוסטט כדי להכיל את דרישות ה-I/O הגדולות יותר בתוך אותו טביעת רגל של חיווט.
- **Eagle**: משפחת מעבד Eagle שילבה טכנולוגיות אריזה מדרגיות יותר מהדורות הקודמים. בפרט, אותות עברו דרך שכבות שבב מרובות, מה שאפשר I/O בצפיפות גבוהה מבלי לוותר על ביצועים.
