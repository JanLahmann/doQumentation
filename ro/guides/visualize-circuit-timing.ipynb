Pe lângă [vizualizarea instrucțiunilor pe un Circuit](/guides/visualize-circuits), s-ar putea să vrei să vizualizezi programarea unui Circuit folosind metoda Qiskit [`timeline_drawer`](https://docs.quantum.ibm.com/api/qiskit/qiskit.visualization.timeline_drawer). Această vizualizare te-ar putea ajuta să identifici rapid timpii de inactivitate ai Qubit-urilor, de exemplu. Totuși, această metodă nu returnează rezultate corecte pentru circuitele dinamice. Pentru a vizualiza programarea circuitelor dinamice, folosește metoda `draw_circuit_schedule_timing`, după cum este descris în secțiunea [Suport Qiskit Runtime](#qr-support).

## Exemple

Pentru a vizualiza un program Circuit programat, poți apela această funcție cu un set de argumente de control. Cea mai mare parte a aspectului imaginii de ieșire poate fi modificată printr-o foaie de stiluri, dar aceasta nu este obligatorie.

### Desenează cu foaia de stiluri implicită

````python
from qiskit import QuantumCircuit
from qiskit.visualization.timeline import draw
from qiskit.providers.fake_provider import GenericBackendV2
from qiskit.transpiler import generate_preset_pass_manager

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)

backend = GenericBackendV2(5)

pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_circuit = pm.run(qc)

draw(isa_circuit, target=backend.target)
````

![Ieșirea celulei de cod anterioare](../docs/images/guides/visualize-circuit-timing/extracted-outputs/5b21b8c1-0.svg)

### Desenează cu o foaie de stiluri potrivită pentru depanarea programelor

````python
from qiskit import QuantumCircuit
from qiskit.visualization.timeline import draw, IQXDebugging
from qiskit.providers.fake_provider import GenericBackendV2
from qiskit.transpiler import generate_preset_pass_manager

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

backend = GenericBackendV2(5)
pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_circuit = pm.run(qc)
draw(isa_circuit, style=IQXDebugging(), target=backend.target)
````

![Ieșirea celulei de cod anterioare](../docs/images/guides/visualize-circuit-timing/extracted-outputs/907dc46c-0.svg)

Poți crea funcții personalizate de generare sau de aspect și actualiza o foaie de stiluri existentă cu funcțiile personalizate. Astfel, poți controla cea mai mare parte a aspectului imaginii de ieșire fără a modifica baza de cod a desenatorului de circuite programate. Consultă referința API [`timeline_drawer`](https://docs.quantum.ibm.com/api/qiskit/qiskit.visualization.timeline_drawer) pentru mai multe exemple.

## Suport Qiskit Runtime
Deși desenatorul de cronologie integrat în Qiskit este util pentru Circuit-uri statice, este posibil să nu reflecte cu acuratețe cronologia [circuit-urilor dinamice](/guides/classical-feedforward-and-control-flow) din cauza operațiunilor implicite, cum ar fi difuzarea și determinarea ramurilor. Ca parte a suportului pentru circuit-uri dinamice, Qiskit Runtime returnează informațiile precise despre cronologia Circuit-ului în rezultatele jobului atunci când este solicitat.

> **Note:** - Aceasta este o funcție experimentală. Se află în stadiu de previzualizare și poate suferi modificări.
> - Această funcție se aplică doar joburilor Qiskit Runtime Sampler.
> - Deși timpul total al Circuit-ului este returnat în metadatele „compilation", acesta NU este timpul utilizat pentru facturare (quantum time).
### Activarea recuperării datelor de cronologie
Pentru a activa recuperarea datelor de cronologie, setează indicatorul experimental `scheduler_timing` la `True` când rulezi jobul primitiv.
