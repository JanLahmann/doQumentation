Tłumienie błędów odnosi się do technik, w których wykorzystujesz wiedzę o niepożądanych efektach, aby wprowadzić dostosowania, które mogą przewidzieć i uniknąć potencjalnych skutków tych efektów. Techniki te często polegają na modyfikowaniu lub dodawaniu sygnałów sterujących, aby procesor kwantowy zwracał oczekiwane wyniki. Zazwyczaj wiąże się to z narzutem w zakresie wstępnego przetwarzania kwantowego; dlatego ważne jest, aby zachować równowagę między doskonaleniem wyników a zapewnieniem, że zadanie zostanie ukończone w rozsądnym czasie.

Prymitywy obsługują wiele technik tłumienia błędów, w tym [dynamiczne odsprzęganie](https://docs.quantum.ibm.com/api/qiskit-ibm-runtime/options-dynamical-decoupling-options#dynamicaldecouplingoptions) i [splatanie Pauliego](https://docs.quantum.ibm.com/api/qiskit-ibm-runtime/options-twirling-options). Wyjaśnienie każdej z nich znajdziesz w artykule [Techniki mitygacji i tłumienia błędów](error-mitigation-and-suppression-techniques). Używając prymitywów, możesz włączać i wyłączać poszczególne metody. Szczegóły znajdziesz w sekcji [Zaawansowane opcje tłumienia błędów](#transpilation-table).

## Zaawansowane opcje tłumienia błędów
W prymitywach możesz jawnie włączać i wyłączać poszczególne metody mitygacji i tłumienia błędów, takie jak dynamiczne odsprzęganie.

> **Note:** - Nie wszystkie opcje są dostępne dla obu prymitywów. Lista dostępnych opcji znajduje się w tabeli [dostępnych opcji](runtime-options-overview#options-table).

````python
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit_ibm_runtime import SamplerV2 as Sampler

service = QiskitRuntimeService()
backend = service.least_busy(operational=True, simulator=False)

sampler = Sampler(backend)

# Turn on dynamical decoupling with sequence XpXm.
sampler.options.dynamical_decoupling.enable = True
sampler.options.dynamical_decoupling.sequence_type = "XpXm"

print(
    f">>> dynamical decoupling sequence to use: {sampler.options.dynamical_decoupling.sequence_type}"
)
````
