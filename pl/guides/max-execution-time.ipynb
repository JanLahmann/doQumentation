## Ustawianie maksymalnego czasu wykonania
Maksymalny czas wykonania zadania to mniejsza z poniższych wartości:

- Wartość ustawiona dla `max_execution_time`
- Wartość limitu czasu zadania określona przez usługę

Wartość `max_execution_time` opiera się na _czasie kwantowym_, a nie na czasie zegarowym. Czas kwantowy to czas, przez który QPU jest dedykowany do przetwarzania twojego zadania.

Ustaw maksymalny czas wykonania (w sekundach) w opcjach zadania, jak pokazano w poniższym przykładzie. Aby uzyskać informacje o ustawianiu opcji, zapoznaj się z artykułem [Specify options](/guides/specify-runtime-options).
