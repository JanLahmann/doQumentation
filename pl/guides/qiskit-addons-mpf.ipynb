Formuły wieloproduktowe (MPF) umożliwiają dokładniejszą symulację dynamiki układu kwantowego kosztem zwiększonej liczby wykonań Circuit. Jest to technika post-processingowa, która łagodzi błąd wartości oczekiwanych dla stanów poddanych ewolucji czasowej.

Obliczenia klasyczne są używane do rozwiązania układu równań liniowych, który dostarcza współczynników do ważonej kombinacji kilku wykonań Circuit. Użycie tej ważonej kombinacji może zmniejszyć błąd związany z symulacją ewolucji czasowej, pod warunkiem dobrego doboru kroków Trottera. Narzędzie MPF przyjmuje zestaw danych — w tym liczbę kroków Trottera i rząd przybliżenia Trottera — aby przygotować i rozwiązać (lub przybliżyć rozwiązanie) powiązanego układu równań liniowych, który następnie możesz użyć do post-processingu pomiarów wartości oczekiwanych stanu poddanego ewolucji czasowej.
