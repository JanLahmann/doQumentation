# Hello World: Circuit Quantum Pertama Anda

Bina sebuah [Bell state](https://en.wikipedia.org/wiki/Bell_state) (dua qubit yang terjerat) dan jalankan dengan tiga cara:

1. **Simulasi ideal** — hasil sempurna, tiada akaun diperlukan
2. **Simulasi berbising** — mensimulasikan peranti sebenar, tiada akaun diperlukan
3. **Perkakasan quantum sebenar** — memerlukan akaun IBM Quantum percuma (langkah persediaan di bawah)

## Bina Circuit

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

qc.draw(output="mpl")
```

## Pilihan 1: Simulasi ideal (tiada akaun diperlukan)
Menggunakan `StatevectorSampler` — simulator tempatan dengan hasil yang sempurna dan bebas bunyi.

```python
from qiskit.primitives import StatevectorSampler

result = StatevectorSampler().run([qc], shots=1024).result()
counts = result[0].data.meas.get_counts()
counts
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

## Pilihan 2: Simulasi berbising (tiada akaun diperlukan)
Menggunakan `FakeManilaV2` — simulator tempatan yang meniru peranti quantum IBM sebenar, termasuk ciri-ciri bunyinya. Circuit perlu di-transpile (disesuaikan) terlebih dahulu kepada set gate dan sambungan qubit peranti tersebut.

```python
from qiskit_ibm_runtime import SamplerV2
from qiskit_ibm_runtime.fake_provider import FakeManilaV2
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

backend = FakeManilaV2()
pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_qc = pm.run(qc)

result = SamplerV2(mode=backend).run([isa_qc], shots=1024).result()
counts = result[0].data.meas.get_counts()
counts
```

```python
plot_histogram(counts)
```

## Pilihan 3: Perkakasan quantum sebenar
Memerlukan akaun IBM Quantum percuma. Untuk mendaftarkan satu:

1. Daftar di [quantum.cloud.ibm.com/registration](https://quantum.cloud.ibm.com/registration) — tiada kad kredit diperlukan untuk 30 hari pertama
2. Log masuk di [quantum.cloud.ibm.com](https://quantum.cloud.ibm.com) dan pilih rantau **us-east** (diperlukan untuk Open Plan percuma)
3. Cipta sebuah instance (Open Plan percuma) di [Instances](https://quantum.cloud.ibm.com/instances) jika anda belum memilikinya
4. Cipta API key di [quantum.cloud.ibm.com](https://quantum.cloud.ibm.com) (atau di [cloud.ibm.com/iam/apikeys](https://cloud.ibm.com/iam/apikeys))
5. Salin **CRN** (Cloud Resource Name) anda dari halaman [Instances](https://quantum.cloud.ibm.com/instances) anda

Kemudian, jika anda belum menyimpan kelayakan anda dalam sesi Binder ini, jalankan sel di bawah. Gantikan `<your-api-key>` dengan API key dari langkah 4, dan `<your-crn>` dengan CRN dari langkah 5.

```python
from qiskit_ibm_runtime import QiskitRuntimeService

QiskitRuntimeService.save_account(
    channel="ibm_quantum_platform",
    token="<your-api-key>",
    instance="<your-crn>",
    set_as_default=True,
    overwrite=True,
)
```

**Nota:** Kerja pada perkakasan sebenar mungkin mengambil masa bergantung pada masa dalam giliran. Jika sel masih berjalan, anda boleh menyemak status kerja dan melihat keputusan di [quantum.cloud.ibm.com/workloads](https://quantum.cloud.ibm.com/workloads?user=me).

```python
from qiskit_ibm_runtime import QiskitRuntimeService, SamplerV2
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

service = QiskitRuntimeService()
backend = service.least_busy(operational=True, simulator=False)
print(f"Running on {backend.name}")

pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_qc = pm.run(qc)

result = SamplerV2(mode=backend).run([isa_qc], shots=1024).result()
counts = result[0].data.meas.get_counts()
counts
```

```python
plot_histogram(counts)
```
