Puedes ejecutar tus cargas de trabajo de primitivas de Qiskit usando APIs REST en uno de tres modos de ejecución, según tus necesidades: job, session y batch. Este tema explica estos modos.

> **Note:** Esta documentación utiliza el módulo `requests` de Python para demostrar la API REST de Qiskit Runtime. Sin embargo, este flujo de trabajo puede ejecutarse con cualquier lenguaje o framework que admita el trabajo con APIs REST. Consulta la [documentación de referencia de la API](https://docs.quantum.ibm.com/api/qiskit-ibm-runtime/tags/jobs) para más detalles.

## Modo job con la API REST
En el modo job, se realiza una única solicitud de primitiva del Estimator o el Sampler sin un gestor de contexto. Consulta cómo ejecutar un Circuit cuántico usando [Estimator](primitives-rest-api#estimator-primitive-with-rest-api) y [Sampler](primitives-rest-api#sampler-primitive-with-rest-api) para ver algunos ejemplos.

## Modo session con la API REST
Una sesión es una característica de Qiskit Runtime que te permite ejecutar eficientemente cargas de trabajo iterativas con múltiples jobs en computadoras cuánticas. El uso de sesiones ayuda a evitar los retrasos causados por poner en cola cada job por separado, lo que puede ser especialmente útil para tareas iterativas que requieren comunicación frecuente entre recursos clásicos y cuánticos. Puedes encontrar más detalles sobre las sesiones en la [documentación](./execution-modes).

> **Note:** Los usuarios del plan Open no pueden enviar jobs de sesión.
