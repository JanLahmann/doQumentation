## ตระกูลโปรเซสเซอร์ที่เลิกใช้งานแล้ว
- **Canary**: ตระกูล Canary ประกอบด้วยดีไซน์ขนาดเล็กที่มี Qubit ตั้งแต่ 5 ถึง 16 ตัว
- **Falcon**: ตระกูล Falcon ของอุปกรณ์เสนอแพลตฟอร์มที่มีคุณค่าสำหรับ Circuit ขนาดกลาง และยังทำหน้าที่เป็นแพลตฟอร์มที่มีคุณค่าสำหรับการสาธิตการปรับปรุงประสิทธิภาพและความสามารถในการขยายก่อนที่จะนำไปใช้บนอุปกรณ์ขนาดใหญ่กว่า
- **Egret**: Egret นำนวัตกรรม tunable coupler มาสู่แพลตฟอร์ม 33 Qubit ส่งผลให้ Gate สองQubit ทำงานได้เร็วขึ้นและมีความเที่ยงตรงสูงขึ้น
- **Hummingbird**: โดยใช้เลย์เอาต์ Qubit แบบ heavy-hexagonal ตระกูล Hummingbird รองรับ Qubit ได้สูงสุด 65 ตัว
- **Osprey**: Osprey มีขนาดใหญ่เกือบสี่เท่าของ Eagle ที่ 433 Qubit ขนาดชิปที่ใหญ่ขึ้นต้องการการปรับปรุงเพิ่มเติมในด้านบรรจุภัณฑ์อุปกรณ์ รวมถึงสาย flex แบบกำหนดเองในตู้เย็น cryostat เพื่อรองรับข้อกำหนด I/O ที่มากขึ้นภายในพื้นที่การเดินสายเดิม
- **Eagle**: ตระกูลโปรเซสเซอร์ Eagle ใช้เทคโนโลยีบรรจุภัณฑ์ที่ขยายขนาดได้ดีกว่ารุ่นก่อนหน้า โดยเฉพาะอย่างยิ่ง สัญญาณส่งผ่านชั้นชิปหลายชั้น ทำให้มี I/O ความหนาแน่นสูงโดยไม่สูญเสียประสิทธิภาพ
