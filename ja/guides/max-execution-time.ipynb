# Qiskit Runtimeワークロードの最大実行時間

公平性を確保するために、各Qiskit Runtimeジョブが実行できる時間には[サービスが算出する](#max-QPU)最大値が設定されています。ジョブがこの制限時間を超えた場合、強制的にキャンセルされ、`RuntimeJobMaxTimeoutError`例外が発生します。コストを管理するために、ジョブに[最大実行時間を設定する](#max-job)こともできます。

 > **Note:** コスト上限に達すると、セッションまたはバッチは即座に終了されます。そのセッションまたはバッチで現在実行中のジョブは完了まで実行されますが、保留中のジョブはキャンセルされます。

<span id="max-job"></span>
## 最大実行時間の設定
ジョブの最大実行時間は、以下の値のうち小さい方が適用されます：

- `max_execution_time`に設定された値
- サービスが決定するジョブのタイムアウト値

`max_execution_time`の値は、ウォールクロック時間ではなく、*QPU使用量*に基づいています。

以下の例に示すように、ジョブオプションで最大実行時間（秒単位）を設定してください。オプションの設定方法については、[Estimatorオプション](/guides/estimator-options)、[Samplerオプション](/guides/sampler-options)、または[Executorオプション](/guides/executor-options)ガイドを参照してください。
