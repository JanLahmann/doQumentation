## Überblick
Fire Opal Performance Management macht es für jeden einfach, mit Quantencomputern in großem Maßstab sinnvolle Ergebnisse zu erzielen – ohne Experte für Quantenhardware sein zu müssen. Beim Ausführen von Circuits mit Fire Opal Performance Management werden KI-gestützte Fehlerunterdrückungstechniken automatisch angewendet, sodass größere Probleme mit mehr Gates und Qubits skaliert werden können. Dieser Ansatz reduziert die Anzahl der benötigten Shots, um die richtige Antwort zu erhalten, ohne zusätzlichen Overhead – was zu erheblichen Einsparungen bei Rechenzeit und Kosten führt.

Performance Management unterdrückt Fehler und erhöht die Wahrscheinlichkeit, auf rauschbehafteter Hardware die richtige Antwort zu erhalten. Mit anderen Worten: Es verbessert das Signal-Rausch-Verhältnis. Das folgende Bild zeigt, wie die durch Performance Management ermöglichte höhere Genauigkeit den Bedarf an zusätzlichen Shots im Fall eines 10-Qubit Quantum Fourier Transform-Algorithmus reduzieren kann. Mit nur 30 Shots erreicht Q-CTRL die 99%-Konfidenzschwelle, während die Standardkonfiguration (`QiskitRuntime` Sampler, `optimization_level`=3 und `resilience_level`=1, `ibm_sherbrooke`) 170.000 Shots benötigt. Durch die schnellere Ermittlung der richtigen Antwort sparst du erhebliche Rechenzeit.

![Visualisierung der verbesserten Laufzeit](../docs/images/guides/qctrl-performance-management/achieve_more.svg)

Die Performance Management Function kann für jeden Algorithmus verwendet werden und lässt sich problemlos an Stelle der Standard-[Qiskit Runtime Primitives](/guides/primitives) einsetzen. Im Hintergrund arbeiten mehrere Fehlerunterdrückungstechniken zusammen, um Fehler zur Laufzeit zu verhindern. Alle Fire Opal Pipeline-Methoden sind vorkonfiguriert und algorithmusunabhängig, sodass du stets die beste Leistung sofort nach dem Start erhältst.

## Beschreibung
Fire Opal Performance Management bietet zwei Ausführungsoptionen, die den Qiskit Runtime Primitives ähneln, sodass du den Q-CTRL Sampler und Estimator einfach austauschen kannst. Der allgemeine Arbeitsablauf bei der Verwendung der Performance Management Function ist:
1. Definiere deinen Circuit (und Operatoren im Fall des Estimators).
2. Führe den Circuit aus.
3. Rufe die Ergebnisse ab.

Um Hardware-Rauschen zu reduzieren, setzt Fire Opal eine Reihe von KI-gesteuerten Fehlerunterdrückungstechniken ein, die im folgenden Bild dargestellt sind. Mit Fire Opal ist die gesamte Pipeline vollständig automatisiert und erfordert keinerlei Konfiguration.

Fire Opals Pipeline beseitigt den Bedarf an zusätzlichem Overhead, wie erhöhter Quantenlaufzeit oder zusätzlichen physischen Qubits. Beachte, dass die klassische Verarbeitungszeit weiterhin ein Faktor ist (siehe den Abschnitt [Benchmarks](#benchmarks) für Schätzwerte, wobei „Gesamtzeit" sowohl klassische als auch Quantenverarbeitung umfasst). Im Gegensatz zur Fehlerminderung, die Overhead in Form von Sampling erfordert, arbeitet Fire Opals Fehlerunterdrückung sowohl auf Gate- als auch auf Pulsebene, um verschiedene Rauschquellen anzugehen und die Wahrscheinlichkeit eines Fehlers zu verringern. Durch die Verhinderung von Fehlern entfällt der Bedarf an aufwändiger Nachverarbeitung.

Das folgende Bild zeigt die von Fire Opal Performance Management automatisierten Fehlerunterdrückungsmethoden.

![Visualisierung der Fehlerunterdrückungspipeline](../docs/images/guides/qctrl-performance-management/error_suppression.svg)

Die Function bietet zwei Primitives – Sampler und Estimator – und die Ein- und Ausgaben beider erweitern die implementierte Spezifikation für [Qiskit Runtime V2 Primitives](/guides/primitive-input-output#pubs).

Die folgende Tabelle bietet eine grobe Orientierung zu Genauigkeit und Laufzeiten aus früheren Benchmarking-Läufen auf `ibm_fez`. Die Leistung auf anderen Geräten kann variieren. Die Nutzungszeit basiert auf der Annahme von 10.000 Shots pro Circuit. Die angegebene „Anzahl der Qubits" ist keine harte Begrenzung, sondern stellt grobe Schwellenwerte dar, bei denen du eine äußerst konsistente Lösungsgenauigkeit erwarten kannst. Größere Problemgrößen wurden erfolgreich gelöst, und Tests jenseits dieser Grenzen sind ausdrücklich empfohlen.

| Beispiel    | Anzahl der Qubits | Genauigkeit | Genauigkeitsmaß | Gesamtzeit (s) | Laufzeitnutzung (s) | Primitive (Modus) |
| ---------  | ---------------- | -------------------------- | -------- | ---------- | ------------- |------------- |
| Bernstein–Vazirani  |  50Q    | 100%  | Erfolgsrate (Anteil der Läufe, bei denen die richtige Antwort den höchsten Zählwert hat)     | 10    | 8         | Sampler |
| Quantum Fourier Transform | 30Q              | 100% | Erfolgsrate (Anteil der Läufe, bei denen die richtige Antwort den höchsten Zählwert hat)      | 10    | 8        | Sampler |
| Quantum Phase Estimation  | 30Q   | 99,9998%  | Genauigkeit des gefundenen Winkels: `1- abs(real_angle - angle_found)/pi`  | 10  | 8  | Sampler |
| Quantensimulation: Ising-Modell (15 Schritte) | 20Q  | 99,775%   |  $A$ (unten definiert)  |  60 (pro Schritt)  | 15 (pro Schritt)   | Estimator |
| Quantensimulation 2: Molekulardynamik (20 Zeitpunkte) | 34Q  |  96,78%  |  $A_{mean}$ (unten definiert)   | 10 (pro Zeitpunkt)    | 6 (pro Zeitpunkt)  | Estimator |

Definition der Genauigkeit der Messung eines Erwartungswerts – die Metrik $A$ ist wie folgt definiert:
$$
A = 1 - \frac{|\epsilon^{ideal} - \epsilon^{meas}|}{\epsilon^{ideal}_{max} - \epsilon^{ideal}_{min}},
$$
wobei $ \epsilon^{ideal} $ = idealer Erwartungswert,  $ \epsilon^{meas} $ = gemessener Erwartungswert, $\epsilon^{ideal}_{max} $ = idealer Maximalwert und $\epsilon^{ideal}_{min}$ = idealer Minimalwert. $A_{mean}$ ist schlicht der Durchschnitt des Werts $A$ über mehrere Messungen.

Diese Metrik wird verwendet, weil sie gegenüber globalen Verschiebungen und Skalierungen im Bereich der erreichbaren Werte invariant ist. Anders ausgedrückt: Egal ob du den Bereich möglicher Erwartungswerte nach oben oder unten verschiebst oder die Streuung erhöhst – der Wert von $A$ bleibt konsistent.

## Estimator primitive

### Estimator example

Use Fire Opal Performance Management's Estimator primitive to determine the expectation value of a single circuit-observable pair.

In addition to the `qiskit-ibm-catalog` and `qiskit` packages, you will also use the `numpy` package to run this example. You can install this package by uncommenting the following cell if you are running this example in a notebook using the IPython kernel.

**2. Circuit ausführen**

Führe den Circuit aus und definiere optional das Backend und die Anzahl der Shots.

````python
import numpy as np
from qiskit.circuit.library import iqp
from qiskit.quantum_info import random_hermitian, SparsePauliOp

n_qubits = 50

# Generate a random circuit
mat = np.real(random_hermitian(n_qubits, seed=1234))
circuit = iqp(mat)
circuit.measure_all()

# Define observables as a string
observable = SparsePauliOp("Z" * n_qubits)
````

````python
# Create PUB tuple
estimator_pubs = [(circuit, observable)]
````

````python
# Run the circuit using Estimator
qctrl_estimator_job = perf_mgmt.run(
    primitive="estimator",
    pubs=estimator_pubs,
    backend_name=backend_name,
)
````

## Sampler Primitive
### Sampler-Beispiel
Verwende das Sampler Primitive von Fire Opal Performance Management, um einen Bernstein–Vazirani-Circuit auszuführen. Dieser Algorithmus, der verwendet wird, um einen verborgenen String aus den Ausgaben einer Black-Box-Funktion zu finden, ist ein gängiger Benchmarking-Algorithmus, da es genau eine richtige Antwort gibt.
**1. Circuit erstellen**

Definiere die richtige Antwort auf den Algorithmus, den verborgenen Bitstring und den Bernstein–Vazirani-Circuit. Du kannst die Breite des Circuits einfach durch Ändern von `circuit_width` anpassen.

````python
import qiskit

circuit_width = 35
hidden_bitstring = "1" * circuit_width

# Create circuit, reserving one qubit for BV oracle
bv_circuit = qiskit.QuantumCircuit(circuit_width + 1, circuit_width)
bv_circuit.x(circuit_width)
bv_circuit.h(range(circuit_width + 1))
for input_qubit, bit in enumerate(reversed(hidden_bitstring)):
    if bit == "1":
        bv_circuit.cx(input_qubit, circuit_width)
bv_circuit.barrier()
bv_circuit.h(range(circuit_width + 1))
bv_circuit.barrier()
for input_qubit in range(circuit_width):
    bv_circuit.measure(input_qubit, input_qubit)

# Create PUB tuple
sampler_pubs = [(bv_circuit,)]
````

**2. Circuit ausführen**

Führe den Circuit aus und definiere optional das Backend und die Anzahl der Shots.

````python
# Run the circuit using Sampler
qctrl_sampler_job = perf_mgmt.run(
    primitive="sampler",
    pubs=sampler_pubs,
    backend_name=backend_name,
)
````

Prüfe den [Status](/guides/functions#check-job-status) deiner Qiskit Function-Arbeitslast oder rufe [Ergebnisse](/guides/functions#retrieve-results) wie folgt ab:

````python
# Print the ID so you can use it later, if necessary
print(qctrl_sampler_job.job_id)

qctrl_sampler_job.status()
````

**3. Retrieve the result**

````python
# Retrieve the job results
sampler_result = qctrl_sampler_job.result()
````

````python
# Get results for the first (and only) PUB
pub_result = sampler_result[0]
counts = pub_result.data.c.get_counts()

print("Counts for the meas output register (limited to 30 results):")
for i, (bitstring, count) in enumerate(counts.items()):
    if i >= 50:
        print(f"  ... ({len(counts) - 30} more items)")
        break
    print(f"  {bitstring}: {count}")
````

**3. Plot the top bitstrings**

Plot the bitstring with the highest counts to see if the hidden bitstring was the mode.

````python
import matplotlib.pyplot as plt


def plot_top_bitstrings(counts_dict, hidden_bitstring=None):
    # Sort and take the top 100 bitstrings
    top_100 = sorted(counts_dict.items(), key=lambda x: x[1], reverse=True)[
        :100
    ]
    if not top_100:
        print("No bitstrings found in the input dictionary.")
        return

    # Unzip the bitstrings and their counts
    bitstrings, counts = zip(*top_100)

    # Assign colors: purple if the bitstring matches hidden_bitstring,
    # otherwise gray
    colors = [
        "#680CE9" if bit == hidden_bitstring else "gray" for bit in bitstrings
    ]

    # Create the bar plot
    plt.figure(figsize=(15, 8))
    plt.bar(
        range(len(bitstrings)), counts, tick_label=bitstrings, color=colors
    )

    # Rotate the bitstrings for better readability
    plt.xticks(rotation=90, fontsize=8)
    plt.xlabel("Bitstrings")
    plt.ylabel("Counts")
    plt.title("Top 100 Bitstrings by Counts")

    # Show the plot
    plt.tight_layout()
    plt.show()
````

The hidden bitstring is highlighted in purple, and it should be the bitstring with the highest number of counts.

````python
plot_top_bitstrings(counts, hidden_bitstring)
````

**3. Die häufigsten Bitstrings darstellen**

Stelle den Bitstring mit den höchsten Zählwerten dar, um zu prüfen, ob der verborgene Bitstring der Modalwert war.
