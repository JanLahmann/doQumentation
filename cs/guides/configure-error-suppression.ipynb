Potlačení chyb označuje techniky, při nichž využíváš znalosti o nežádoucích efektech k zavedení přizpůsobení, které dokáže tyto efekty předvídat a omezit jejich možné dopady. Tyto techniky často spočívají ve změně nebo přidání řídicích signálů, aby kvantový procesor vrátil požadované výsledky. To obvykle vede k režijním nákladům na kvantové předzpracování; je proto důležité najít rovnováhu mezi zdokonalováním výsledků a zajištěním toho, aby tvoje úloha byla dokončena v přiměřeném čase.

Primitivy podporují řadu technik potlačení chyb, včetně [dynamického oddělování](https://docs.quantum.ibm.com/api/qiskit-ibm-runtime/options-dynamical-decoupling-options#dynamicaldecouplingoptions) a [twirling Pauliho operátorů](https://docs.quantum.ibm.com/api/qiskit-ibm-runtime/options-twirling-options). Vysvětlení každé z nich najdeš v části [Techniky zmírňování a potlačení chyb](error-mitigation-and-suppression-techniques). Při použití primitiv můžeš jednotlivé metody zapínat a vypínat. Podrobnosti najdeš v části [Pokročilé možnosti potlačení chyb](#transpilation-table).

Estimator ve výchozím nastavení používá potlačení a zmírňování chyb. Pokud nechceš, aby se tvé vstupní Circuit provedly jakékoli zpracování, postupuj podle pokynů v části [Vypnutí veškerého zmírňování a potlačení chyb](specify-runtime-options#no-error-mitigation).

## Pokročilé možnosti potlačení chyb
V primitivách můžeš explicitně zapínat a vypínat jednotlivé metody zmírňování a potlačení chyb, například dynamické oddělování.

> **Note:** - Ne všechny možnosti jsou dostupné pro obě primitivy. Seznam dostupných možností najdeš v tabulce [dostupných možností](runtime-options-overview#options-table).
> - Ne všechny metody fungují společně na všech typech Circuit. Podrobnosti najdeš v tabulce [kompatibility funkcí](runtime-options-overview#options-compatibility-table).

````python
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit_ibm_runtime import SamplerV2 as Sampler

service = QiskitRuntimeService()
backend = service.least_busy(operational=True, simulator=False)

sampler = Sampler(backend)

# Turn on dynamical decoupling with sequence XpXm.
sampler.options.dynamical_decoupling.enable = True
sampler.options.dynamical_decoupling.sequence_type = "XpXm"

print(
    f">>> dynamical decoupling sequence to use: {sampler.options.dynamical_decoupling.sequence_type}"
)
````
