## Příklady

Pro vizualizaci naplánovaného programu Circuit můžeš tuto funkci zavolat se sadou řídicích argumentů. Většinu vzhledu výstupního obrázku lze upravit pomocí stylesheetu, ale to není povinné.

### Kreslení s výchozím stylesheetem

````python
from qiskit import QuantumCircuit
from qiskit.visualization.timeline import draw
from qiskit.providers.fake_provider import GenericBackendV2
from qiskit.transpiler import generate_preset_pass_manager

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)

backend = GenericBackendV2(5)

pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_circuit = pm.run(qc)

draw(isa_circuit, target=backend.target)
````

![Výstup předchozí buňky kódu](../docs/images/guides/visualize-circuit-timing/extracted-outputs/5b21b8c1-0.svg)

### Kreslení se stylesheetem vhodným pro ladění programu

````python
from qiskit import QuantumCircuit
from qiskit.visualization.timeline import draw, IQXDebugging
from qiskit.providers.fake_provider import GenericBackendV2
from qiskit.transpiler import generate_preset_pass_manager

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

backend = GenericBackendV2(5)
pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_circuit = pm.run(qc)
draw(isa_circuit, style=IQXDebugging(), target=backend.target)
````

![Výstup předchozí buňky kódu](../docs/images/guides/visualize-circuit-timing/extracted-outputs/907dc46c-0.svg)

Můžeš vytvořit vlastní generátorové nebo layoutové funkce a aktualizovat jimi existující stylesheet. Tímto způsobem lze ovládat většinu vzhledu výstupního obrázku bez nutnosti upravovat zdrojový kód kreslítka naplánovaných Circuit. Další příklady najdeš v referenci API [`timeline_drawer`](https://docs.quantum.ibm.com/api/qiskit/qiskit.visualization.timeline_drawer).

## Podpora Qiskit Runtime
Přestože je časová osa (timeline drawer) zabudovaná v Qiskitu užitečná pro statické Circuit, nemusí přesně odrážet časování [dynamických Circuit](/guides/classical-feedforward-and-control-flow), a to kvůli implicitním operacím, jako je broadcasting a určování větví. Jako součást podpory dynamických Circuit vrací Qiskit Runtime přesné informace o časování Circuit uvnitř výsledků úlohy – pokud jsou vyžádány.

> **Note:** - Jedná se o experimentální funkci. Je ve stavu preview release, a proto se může změnit.
> - Tato funkce se vztahuje pouze na úlohy Qiskit Runtime Sampler.
> - Ačkoli je celkový čas Circuit vrácen v metadatech „compilation", NENÍ to čas používaný pro fakturaci (kvantový čas).

### Povolení načítání dat o časování
Chceš-li povolit načítání dat o časování, nastav experimentální příznak `scheduler_timing` na `True` při spuštění primitivní úlohy.
