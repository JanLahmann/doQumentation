## Režim job s REST API
V režimu job se provede jediný primitivní požadavek Estimatoru nebo Sampleru bez správce kontextu. Příklady spuštění kvantového Circuit najdeš v části [Estimator](primitives-rest-api#estimator-primitive-with-rest-api) a [Sampler](primitives-rest-api#sampler-primitive-with-rest-api).

## Režim session s REST API
Session je funkce Qiskit Runtime, která ti umožňuje efektivně spouštět iterativní workloady s více úlohami na kvantových počítačích. Používání sessions pomáhá předcházet zpožděním způsobeným samostatným zařazováním každé úlohy do fronty, což může být zvláště užitečné pro iterativní úlohy vyžadující časté přepínání mezi klasickými a kvantovými zdroji. Další podrobnosti o Sessions najdeš v [dokumentaci](./execution-modes).

> **Note:** Uživatelé plánu Open nemohou odesílat úlohy v session.
