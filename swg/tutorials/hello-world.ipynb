# Hello World: Euer erschter Quanteschaltkreis

Bauet an [Bell-Zustand](https://en.wikipedia.org/wiki/Bell_state) (zwoi verschränkte Qubits) und führet en auf drei Arte aus:

1. **Ideali Simulation** — perfekte Ergebnis, koi Konto braucht mr ned
2. **Simulation mit Rausche** — simuliert a echts Gerät, koi Konto braucht mr ned
3. **Echti Quantehardware** — bruucht a [IBM Quantum-Konto](https://janlahmann.github.io/Qiskit-documentation/#setting-up-your-ibm-quantum-account)

## Schaltkreis erstelle

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

qc.draw(output="mpl")
```

## Option 1: Ideali Simulation (koi Konto nötig)
Verwendet `StatevectorSampler` — a lokaler Simulator mit perfekte, rauschfreie Ergebnis.

```python
from qiskit.primitives import StatevectorSampler

result = StatevectorSampler().run([qc], shots=1024).result()
counts = result[0].data.meas.get_counts()
counts
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

## Option 2: Simulation mit Rausche (koi Konto nötig)
Verwendet `FakeManilaV2` — a lokaler Simulator, wo a echts IBM-Quantegerät nachahmt, au mit seine Rauscheigenschafte. Dr Schaltkreis muess zerscht transpiliert (angepasst) werre, damit er zum Gate-Satz und zur Qubit-Konnektivität vom Gerät passt.

```python
from qiskit_ibm_runtime import SamplerV2
from qiskit_ibm_runtime.fake_provider import FakeManilaV2
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

backend = FakeManilaV2()
pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_qc = pm.run(qc)

result = SamplerV2(mode=backend).run([isa_qc], shots=1024).result()
counts = result[0].data.meas.get_counts()
counts
```

```python
plot_histogram(counts)
```

## Option 3: Echti Quantehardware
Bruucht a IBM Quantum-Konto. Gucket Se uf [Setting up your IBM Quantum account](https://janlahmann.github.io/Qiskit-documentation/#setting-up-your-ibm-quantum-account) für d' Details.

Falls Se Ihre Zugangsdaten no net in derer Binder-Sitzung gspeichert henn, führet Se des zerscht aus:

```python
from qiskit_ibm_runtime import QiskitRuntimeService

QiskitRuntimeService.save_account(
    token="<your-api-key>",
    instance="<your-crn>",
    overwrite=True
)
```

**Hinweis:** Jobs uf echter Hardware ka's je nach Wartezeite in dr Warteschlange a Weilele dauere. Falls d' Zelle no lauft, ka mr de Job-Status gucke und d' Ergebnis uf [quantum.cloud.ibm.com/workloads](https://quantum.cloud.ibm.com/workloads?user=me) aluege.

```python
from qiskit_ibm_runtime import QiskitRuntimeService, SamplerV2
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

service = QiskitRuntimeService()
backend = service.least_busy(operational=True, simulator=False)
print(f"Running on {backend.name}")

pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_qc = pm.run(qc)

result = SamplerV2(mode=backend).run([isa_qc], shots=1024).result()
counts = result[0].data.meas.get_counts()
counts
```

```python
plot_histogram(counts)
```
